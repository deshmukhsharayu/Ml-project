```python
#Import necessary labraries

import pandas as pd
```

```python
#collect Dataset

import kagglehub

# Download latest version
path = kagglehub.dataset_download("shahriarkabir/procurement-strategy-dataset-for-kraljic-matrix")

print("Path to dataset files:", path)
```

```python
#Load the Dataset

df = pd.read_csv("/kaggle/input/procurement-strategy-dataset-for-kraljic-matrix/realistic_kraljic_dataset.csv")
df
```

```python
#Check column

df.columns
```

```python
#Check information

df.info()
```

```python
#Clean the data (Remove/Drop)

df.drop(["Product_Name","Supplier_Region"],axis=1,inplace=True)
```

```python
#Clean the data (Remove/Drop)

df.drop(["Product_ID"],axis=1,inplace=True)
```

```python
df
```

```python
#Import Labraries

import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
#Check column

df.columns
```

```python
df['Single_Source_Risk'] = df['Single_Source_Risk'].map({'Yes':1,'No':0})
```

```python
#Split features x

x = df[['Lead_Time_Days', 'Order_Volume_Units', 'Cost_per_Unit',
       'Supply_Risk_Score', 'Profit_Impact_Score', 'Environmental_Impact',
       'Single_Source_Risk']]
```

```python
#Split features y

y = df['Kraljic_Category']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 ,random_state =12)
```

```python
df
```

```python
x_train
```

```python
y_train
```

```python
#Train different models

model_knn = KNeighborsClassifier()
model_NB  = GaussianNB()
model_LR  = LogisticRegression()
model_svm = SVC()
```

```python
#Evaluation of models in X

model_knn.fit(x_train,y_train)
model_NB.fit(x_train,y_train)
model_LR.fit(x_train,y_train)
model_svm.fit(x_train,y_train)
```

```python
#Evaluation of models in Y

y_pred_knn = model_knn.predict(x_test)
y_pred_NB = model_NB.predict(x_test)
y_pred_LR = model_LR.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
```

```python
#Print evalution result from each model (knn)

print(classification_report(y_test,y_pred_knn))
```

```python
#Print evalution result from each model (NB)

print(classification_report(y_test,y_pred_NB))
```

```python
#Print evalution result from each model (LR)

print(classification_report(y_test,y_pred_LR))
```

```python
#Print evalution result from each model (svm)

print(classification_report(y_test,y_pred_svm))
```
